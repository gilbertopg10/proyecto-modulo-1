2. Implementa el método combate que tome como argumento un rival (otra instancia de la clase Pokemon). El combate va a ser como el de los videojuegos, por lo que va a ser por turnos en los que cada turno cada pokemon realiza un ataque que resta puntos de vida del rival hasta. Este se detiene cuando los puntos de vida de alguno de los dos llegue a cero. Asegúrate que cumpla lo siguiente:

i. El Pokémon de mayor velocidad es el que ataca primero.
ii. Si alguien es atacado por una de sus debilidades, recibe el doble de daño. Si es atacado por una de sus fortalezas, recibe sólo la mitad de daño.
iii. Calcula el daño y réstalos de los puntos de vida.
iv. En cuanto uno tenga 0 puntos de vida, termina el ciclo y declara un ganador


```python
import random

class Pokemon:
    def __init__(self, especie, tipo, fortalezas, debilidades, ataque, defensa, velocidad, puntos_de_vida):
        self.especie = especie
        self.tipo = tipo
        self.fortalezas = fortalezas
        self.debilidades = debilidades
        self.ataque = ataque
        self.defensa = defensa
        self.velocidad = velocidad
        self.puntos_de_vida = puntos_de_vida

    def combate(self, pkm1, pkm2):
        print(f"¡Comienza la batalla entre {pkm1.especie} y {pkm2.especie}!")

        # El Pokémon más rápido ataca primero
        if pkm1.velocidad > pkm2.velocidad:
            atacante = pkm1
            defensor = pkm2
        elif pkm1.velocidad < pkm2.velocidad:
            atacante = pkm2
            defensor = pkm1
        else:
            # En caso de empate en velocidad, se determina al azar
            if random.random() < 0.5:
                atacante = pkm1
                defensor = pkm2
            else:
                atacante = pkm2
                defensor = pkm1

        # Bucle de combate hasta que alguno de los Pokémon quede sin puntos de vida
        while True:
            print(f"{atacante.especie} ataca a {defensor.especie}!")
            if atacante.tipo in defensor.debilidades:
                danio = atacante.ataque * 2 - defensor.defensa
                print("¡Es una debilidad! El ataque hace un daño crítico.")
            elif atacante.tipo in defensor.fortalezas:
                danio = (atacante.ataque * 0.5 - defensor.defensa) // 1
                print("¡Es una fortaleza! El ataque hace menos daño de lo normal.")
            else:
                danio = atacante.ataque - defensor.defensa
            if danio < 1:
                danio = 1
            defensor.puntos_de_vida -= danio
            print(f"{defensor.especie} pierde {danio} puntos de vida.")
            if defensor.puntos_de_vida <= 0:
                print(f"{defensor.especie} ha sido derrotado.")
                print(f"¡{atacante.especie} gana la batalla!")
                break
            atacante, defensor = defensor, atacante  # Cambiar el rol de atacante y defensor

    
```
